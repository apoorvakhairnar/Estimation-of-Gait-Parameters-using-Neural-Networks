```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load("RNN_initial_results/RNN_combined_history.npz", allow_pickle=True)
RNN_combined = data["history"][0]
data = np.load("GRU_initial_results/GRU_combined_history.npz", allow_pickle=True)
GRU_combined = data["history"][0]

data = np.load("K_fold_RNN_GRU_dict.npz", allow_pickle=True)
K_fold = data["data"][0]
RNN_5_fold_K = K_fold['RNN_5_fold']
RNN_10_fold_K = K_fold['RNN_10_fold']
GRU_5_fold_K = K_fold['GRU_5_fold']
GRU_10_fold_K = K_fold['GRU_10_fold']
```

```python
plt.rcParams.update({'font.size': 15})
plt.figure(figsize=(10, 10))
plt.subplot(211)
plt.plot(RNN_combined['history']['val_loss'][:25], label='No CV')
plt.plot(RNN_5_fold_K['history'][3].history['val_loss'][:25], label='Regular 5-fold')
plt.plot(RNN_10_fold_K['history'][3].history['val_loss'][:25], label='Regular 10-fold')
plt.title('RNN Validation Loss for Different CV Techniques')
plt.xlim(-1, 26)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.subplot(212)
plt.plot(GRU_combined['history']['val_loss'][:25], lw=2, label='No CV')
plt.plot(GRU_5_fold_K['history'][3].history['val_loss'][:25], label='5-fold')
plt.plot(GRU_10_fold_K['history'][3].history['val_loss'][:25], label='10-fold')
plt.title('GRU Validation Loss for Different CV Techniques')
plt.xlim(-1, 26)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.show()
```
