```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load("GRU_combined_history_final.npz", allow_pickle=True)
GRU_combined = data["history"][0]
y_test = GRU_combined['pred'][0]
y_pred = GRU_combined['pred'][1]

plt.rcParams.update({'font.size': 15})
plt.figure(figsize=(20, 5))
plt.subplot(131)
plt.plot(GRU_combined['history']['loss'], label='Training Loss')
plt.plot(GRU_combined['history']['val_loss'], label='Validation Loss')
plt.title('Loss Convergence Curves')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.subplot(132)
plt.plot(GRU_combined['history']['mae'], label='Training MAE')
plt.plot(GRU_combined['history']['val_mae'], label='Validation MAE')
plt.title('MAE Convergence Curves')
plt.xlabel('Epochs')
plt.ylabel('MAE')
plt.legend()
plt.subplot(133)
plt.scatter(y_test, y_test, alpha=0.5, color='#2C3E50', label='True', zorder=2)
plt.scatter(y_test, y_pred, alpha=0.5, color='#FF6E40', label='GRU Prediction', zorder=1)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], '--', color='gray')
plt.title('Predictions on Testing Set')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.legend()

plt.show()
```
